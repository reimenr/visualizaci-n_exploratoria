#### Dos variables numéricas (Año o Trimestre y Precio)

```python
plt.figure(figsize=(10, 6))
media_preu = preu_mes.groupby("Trimestre").Preu.mean()
media_preu.plot(style="--", marker="o", ms=8, color="orange")
for trimestre, preu in media_preu.items():
    plt.text(trimestre, preu, f'{preu:.2f}', ha='right', va='bottom', fontsize=8)
plt.ylabel('Preu')
plt.title('Nom_Districte vs Preu/mes')
plt.savefig('Nom_Districte vs Preu-mes.png')
plt.show()
```

En la grafica se puede observar que para el primer trimestre, los precios tuvieron un promedio de 797.85 € para el segundo trimestre bajaron a 793.56 €, y luego se mantuvieron al alza durante los trimestres 3 y 4 hasta alcnazar el un promedio de 823.89 €.

Durante todo el año hubo una variacion de precios/mes de mas de 300 €

```python
plt.figure(figsize=(10, 6))
preu_m2.groupby("Trimestre").Preu.mean().plot(style="--", marker="o", ms=8, color="green")
plt.ylabel('Preu')
plt.title('Nom_Districte vs Preu/m2')
plt.savefig('Nom_Districte vs Preu-m2.png')
plt.show()
```

En esta grafica se puede observar que el poromedio de los precios por m2 se mantuvieron al alza durante los 4 trimestres ( todo el año), sin embargo en el ultimo trimestre el promedio tuvo una variacion mucho mas alta, llegando a tener precios superiores a los 11.950 €, teniendo en cuenta que el minimo precio durante el año fue de 11.650€ existio una variacion promedio de precios de mas de 300€

#### Tres variables (Barrio o Distrito, Trimestre y Precio)

```python
heatmap_data = preu_mes.pivot_table(values='Preu', index='Nom_Districte', columns='Trimestre', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".3f", linewidths=.1, cbar_kws={'label': 'Preu'})
plt.title('Precio/mes media por Trimestre para cada Distrito')
plt.savefig('Precio-mes media por Trimestre para cada Distrito.png')
plt.show()
```

En esta grafica se reflejan tres variables, el distrito, el trimestre y el precio promedio por cada trimestre. el rango de colores nos muestra un rango de precios entre los 500€ y los 1.300€. lo que nos permite visualizar facilmente cuales fueron los distritos que se mantuvieron los precios, como ciutat vella, Gracia, Eixample, Horta, Nou barris y Sants, mientras que en los distritos que mas hubo variacion de precios a lo largo de los 4 trimestres fue en Les Corts, Sant Andreu, Sant Marti y Sarria.

```python
heatmap_data = preu_m2.pivot_table(values='Preu', index='Nom_Districte', columns='Trimestre', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".3f", linewidths=.1, cbar_kws={'label': 'Preu'})
plt.title('Precio/m2 media por Trimestre para cada Distrito')
plt.savefig('Precio-m2 media por Trimestre para cada Distrito.png')
plt.show()
```

En la grafica podemos observar el precio/m2 por cada distrito a lo largo de 4 trimestres, en algunos distritos se mantuvieron homogeneos los precios durante este periodo de tiempo, pero en otros como Les Corts, Sant Andreu, Sant Marti, Sarria, los precios variaron a lo largo de los 4 trimestres. 

Algunas variaciones se mantuvieron al alza a traves del año como ciutat vella, Horta, mientras que otra fueron mas fluctuantes, como en el caso de Les Corts, Sarria, Sant Marti, Sant Andreu, Eixample.

Propone alguna visualización que creas que puede resultar interesante.

```python
data_for_hist = preu_mes[['Preu', 'Nom_Districte', 'Trimestre']]
plt.figure(figsize=(14, 8))
sns.barplot(x='Nom_Districte', y='Preu', hue='Trimestre', data=data_for_hist, errorbar=None, palette='YlGnBu')
plt.xlabel('Nom_Districte')
plt.ylabel('Preu')
plt.title('Distribución de Precios/mes media por Distrito y Trimestre')
plt.legend(title='Trimestre', bbox_to_anchor=(1, 1))
plt.savefig('Distribución de Precios-mes media por Distrito y Trimestre.png')
plt.show()
```

En esta grafica me resulta interesante por que puedo contrastar facilmente los 4 trimestres para cada distrito, en cuanto a los precios y detectar facilmente las varioaciones de las cuatro columnas, además puedo que los barrios con los precios más costosos es Sarria y Les Corts y los mas económicos son: Nou barris, san andreu y Horta-guinardo.

sin embargo los que tuvieron mayor variacion de precios fueron: Sarria, Les corts, Sant Andreu y Sant Marti. 

El barrio mas estable en precio a lo largo de los 4 trimestres del 2021 fue Sants y Nou barris
